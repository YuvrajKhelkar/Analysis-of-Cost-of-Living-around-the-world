```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Cost_of_Living_Index_by_Country_2024.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
top_10_expensive = df.sort_values(by='Cost of Living Index', ascending=False).head(10)
top_10_expensive
```

```python
print("Top 10 Countries with the Highest Cost of Living:\n", top_10_expensive[['Country', 'Cost of Living Index']])
```

```python
top_10_cheapest = df.sort_values(by='Cost of Living Index', ascending=True).head(10)
top_10_cheapest
```

```python
print("Top 10 Countries with the Lowest Cost of Living:\n", top_10_cheapest[['Country', 'Cost of Living Index']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Plot the top 10 countries with the highest cost of living 

plt.figure(figsize= (10,6))
sns.barplot(x = 'Cost of Living Index', y = 'Country' , data = top_10_expensive)
plt.title('Top 10 Countries with the Highest Cost of Living')
plt.xlabel('Cost of living Index')
plt.ylabel('Country')
plt.show()
```

```python
# Plot the top 10 countries with the lowest cost of living
plt.figure(figsize=(10, 6))
sns.barplot(x='Cost of Living Index', y='Country', data=top_10_cheapest)
plt.title('Top 10 Countries with the Lowest Cost of Living')
plt.xlabel('Cost of Living Index')
plt.ylabel('Country')
plt.show()
```

```python
# Scatter plot to visualize the relationship between Cost of Living Index and Rent Index
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Cost of Living Index', y='Rent Index', data=df)
plt.title('Cost of Living Index vs Rent Index')
plt.xlabel('Cost of Living Index')
plt.ylabel('Rent Index')
plt.show()
```

```python
# Scatter plot to visualize the relationship between Cost of Living Index and Groceries Index
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Cost of Living Index', y='Groceries Index', data=df)
plt.title('Cost of Living Index vs Groceries Index')
plt.xlabel('Cost of Living Index')
plt.ylabel('Groceries Index')
plt.show()
```

```python
# Scatter plot to visualize the relationship between Cost of Living Index and Restaurant Price Index
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Cost of Living Index', y='Restaurant Price Index', data=df)
plt.title('Cost of Living Index vs Restaurant Price Index')
plt.xlabel('Cost of Living Index')
plt.ylabel('Restaurant Price Index')
plt.show()
```

```python
# Scatter plot to visualize the relationship between Local Purchasing Power Index and Cost of Living Index
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Local Purchasing Power Index', y='Cost of Living Index', data=df)
plt.title('Local Purchasing Power Index vs Cost of Living Index')
plt.xlabel('Local Purchasing Power Index')
plt.ylabel('Cost of Living Index')
plt.show()
```

```python
#Compare rankings based on different indices
df['Rank by Rent'] = df['Rent Index'].rank(ascending=False)
df['Rank by Groceries'] = df['Groceries Index'].rank(ascending=False)
df['Rank by Restaurant'] = df['Restaurant Price Index'].rank(ascending=False)
df['Rank by Purchasing Power'] = df['Local Purchasing Power Index'].rank()

# Create a DataFrame for comparison
ranking_comparison = df[['Country', 'Rank', 'Rank by Rent', 'Rank by Groceries', 'Rank by Restaurant', 'Rank by Purchasing Power']]

# Display the comparison
print(ranking_comparison.head(10))
```

```python
# Plot the distribution of the Cost of Living Index
plt.figure(figsize=(10, 6))
sns.histplot(df['Cost of Living Index'], bins=20, kde=True)
plt.title('Distribution of Cost of Living Index')
plt.xlabel('Cost of Living Index')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a DataFrame for comparative boxplots
indices_df = df[['Cost of Living Index', 'Rent Index', 'Groceries Index', 'Restaurant Price Index', 'Local Purchasing Power Index']]

# Plot boxplots for each index
plt.figure(figsize=(12, 8))
sns.boxplot(data=indices_df)
plt.title('Comparative Boxplot of Different Indices')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```

```python
# Create a DataFrame for outlier detection
outliers_df = df[['Country', 'Cost of Living Index', 'Rent Index', 'Groceries Index',
                  'Restaurant Price Index', 'Local Purchasing Power Index']]

# Detect outliers using Z-scores
from scipy.stats import zscore
outliers_df = outliers_df[(zscore(outliers_df[['Cost of Living Index', 'Rent Index', 'Groceries Index',
                                                'Restaurant Price Index', 'Local Purchasing Power Index']]) > 3).any(axis=1)]

# Display outliers
print("Outliers:\n", outliers_df)

```
